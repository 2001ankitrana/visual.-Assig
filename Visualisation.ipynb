1/ Using dataset to generate -------- dimensional speech.

```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np

np.random.seed(30)
data = {
    'x':np.random.uniform(-10,10,300),
    'y':np.random.uniform(-10,10,300),
    'z':np.random.uniform(-10,10,300)
}
df = pd.DataFrame(data)

fig = px.scatter_3d(df,x='x',y='y',z='z',color='z')
fig.show()




```

2/ Using students grades create violin plot-----categories.


```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import pandas as pd

np.random.seed(15)
data = {
    'Grade': np.random.choice(['A','B','C','D'], 100),
    'Score': np.random.randint(50,100,size=100)
}
df = pd.DataFrame(data)
fig = px.violin(df,x='Grade',y='Score',color='Grade',box=True,points='all')
fig.show()
```

```python
# heatmap

import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import pandas as pd

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan','Feb','Mar','Apr','May','Jun'],100),
    'Day': np.random.choice(range(1,31),100),
    'Sales': np.random.randint(100,500,100)
}
df = pd.DataFrame(data)
fig = px.density_heatmap(df,x='Month',y='Day',z='Sales',color_continuous_scale='Viridis')
fig.show()
```

4/ Using given x and y ------function

```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import pandas as pd

x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
x,y = np.meshgrid(x,y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'x': x.flatten(),
    'y': y.flatten(),
    'z': z.flatten()
}
df = pd.DataFrame(data)
fig = px.scatter_3d(df,x='x',y='y',z='z',color='z')
fig.show()
```

5/ Using given dataset ------- the population.

```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import pandas as pd

np.random.seed(25)
data = {
    'Country': ['USA','Canada','UK','Germany','France'],
    'Population': np.random.randint(100,1000,5),
    'GDP': np.random.randint(500,2000,5)
}
df = pd.DataFrame(data)
fig = px.scatter(df,x='Population',y='GDP',size='Population',color='GDP')
fig.update_traces(marker=dict(size=30))
fig.update_layout(scene=dict(xaxis_title='GDP',yaxis_title='Population'))
fig.show()


```

1/ Create bokeh plot displaying sine wave-----sine of x.

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
import numpy as np


x = np.linspace(0,1,10)
y = np.sin(x)

p = bokeh.plotting.figure(title='Sine Wave',x_axis_label='x',y_axis_label='sin(x)')
p.line(x,y,legend_label='sin(x)')
bokeh.plotting.show(p)

```

2/ Create a bokeh scatter plot using--------columns.

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
import pandas as pd

x = [1,2,3,4,5,6,7,8,9,18]
y = [2,4,5,7,6,8,9,18,12,13]

df = pd.DataFrame({'x':x,'y':y})
p = bokeh.plotting.figure(title='Scatter Plot',x_axis_label='x',y_axis_label='y')

p.scatter(df['x'],df['y'])
bokeh.plotting.show(p)

```

3/ Generate bokeh bar chart------dataset.

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.io import show, output_notebook
import pandas as pd


fruits = ['Apples','Oranges','Bananas','Pears']
counts = [20,25,30,35]

df = pd.DataFrame({'fruits':fruits,'counts':counts})
p = bokeh.plotting.figure(title='Bar Chart',x_axis_label='fruits',y_axis_label='counts',x_range=fruits)

p.vbar(x=df['fruits'],top=df['counts'])
bokeh.io.show(p)
```

4/ Create bokeh histogram

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
import pandas as pd
import numpy as np
import numpy as ap

data_hist = np.random.randn(1000)
hist,edges = np.histogram(data_hist,bins=50,density=True)

p = bokeh.plotting.figure(title='Histogram',x_axis_label='Value',y_axis_label='Frequency')
p.quad(top=hist,bottom=0,left=edges[:-1],right=edges[1:],fill_color='skyblue',line_color='black')
bokeh.io.show(p)
```

5/ Create bokeh heatmap---- dataset.


```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
import pandas as pd
import numpy as np

data_heatmap = np.random.rand(10,10)
x = np.linspace(0,1,10)
y = np.linspace(0,1,10)
x,y = np.meshgrid(x,y)

p = bokeh.plotting.figure(title='Heatmap',x_axis_label='x',y_axis_label='y')
p.image(image=[data_heatmap],x=[0],y=[0],dw=[1],dh=[1],palette='Viridis256')
bokeh.io.show(p)

```
